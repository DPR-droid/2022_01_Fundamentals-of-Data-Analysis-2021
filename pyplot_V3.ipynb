# 1. Overview

## What is a plot?
A plot is a way to graphically represent a dataset, by using a graph to show the relationship between two or more variables.  A graph can allow a data analyst to visually see the representation of the relationships and derive an understanding which may not have come from a list of values. Plots play an important role in statistics and data analysis. Plots are a short path to gaining insight into a data set in terms of testing assumptions, model selection, model validation, estimator selection, relationship identification, factor effect determination, outlier detection. [1]

## What is Matplotlib? 
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. Matplotlib was originally written by John D. Hunter as an open-source alternative to a tool called MATLAB, a proprietary multi-paradigm programming language and numeric computing environment developed by MathWorks. [3] After 2012 Michael Droettboom was nominated as matplotlib's lead is an active development community and is distributed under a BSD-style license [2]

## What is matplotlib.pyplot? 
As mentioned above Pyplot is a Matplotlib module that provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python and the advantage of being free and open-source. The various plots we can utilize using Pyplot are Line Plot, Histogram, Scatter, 3D Plot, Image, Contour, and Polar. [4]

## Figure Subplots
A matplotlib figure may contain multiple subplots. These subplots are organized in a grid. To create a subplot, just call the subplot function, and specify the number of rows and columns in the figure, and the index of the subplot you want to draw on (starting from 1, then left to right, and top to bottom). [11]

```python
# Create subplots
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottow left
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.savefig("data\subplots.png", transparent=True)
plt.show()
```

The Matplotlib library is extensive and has over 70,000 lines of code [5] There are 2 ways of plotting in matplotlib, those are:
    pyplot API.
    object-oriented

Pyplot API provides the state-machine interface to the underlying object-oriented plotting library. The state-machine implicitly and automatically creates figures and axes to achieve the desired plot. [12]

## Issue with pyplot API? (Do not use it if you are writing a program)
The pyplot API is generally less flexible than the object-oriented API. What does this mean? pyplot API is a state-based interface, where the figure (plot generated) is preserved through various function calls allowing to quickly and easily generate plots. This means that quantities share the same axis but have very different magnitudes, the graphs look disproportionate. The plot approach doesn't scale when we are required to make multiple plots or when we have to make intricate plots that require a lot of customisation. [6]

## 5.1 Scatterplots [15]

A scatter plot is a diagram where each value in the data set is represented by a dot. In Machine Learning the data sets can contain thousands-, or even millions, of values. You might not have real-world data when you are testing an algorithm, you might have to use randomly generated values. Let us create two arrays that are both filled with 1000 random numbers from a normal data distribution.

    The first array will have the mean set to 5.0 with a standard deviation of 1.0.
    The second array will have the mean set to 10.0 with a standard deviation of 2.0:

```python
# Use Numpy random to create arrays for x and y
x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)

# Show scatter plot
plt.scatter(x, y)
plt.savefig("data\scatterplot.png", transparent=True)
plt.show()
```

### Scatter Plot Explained

The above example shows the dots are concentrated around the value 5 on the x-axis and 10 on the y-axis. The spread is wider on the y-axis than on the x-axis. 

### When you should use a scatter plot [19]

Scatter plots’ primary uses are to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points but also patterns when the data are taken as a whole.

Identification of correlational relationships is common with scatter plots. We can divide data points into groups based on how closely sets of points cluster together. 
A scatter plot can also be useful for identifying other patterns in data. Scatter plots can also show if there are any unexpected gaps in the data and if there are any outlier points. 

These are demonstrated in another one of my GitHub projects on the [iris dataset](https://github.com/DPR-droid/pands-project) that uses multiple scatterplots to show the distinct groups among three different species of iris.

## 5.2 Exponential [16]

Most processes in nature are described by exponential functions, If you correctly approximate the available data, then it becomes possible to estimate and predict future values. Exponential growth is an increase in value where the growth rate is proportional to the value of the quantity itself. 

Creating an exponential plot is requires the NumPy library, the mathematical method from the sсipy library, and the matplotlib charting library set above on the imports.

To find the parameters of an exponential function of the form y = a * exp (b * x), we use the optimization method. To do this, the scipy.optimize.curve_fit () the function is suitable for us. 

```python
# numpy to input values for x, y
x = np.arange(1, 31, 1)
y = np.array([3,7,14,16,26,47,73,84,113,196,218,310,356,475,548,645,794,
            942,1096,1251,1319,1462,1668,1892,2203,2511,2777,3102,3372,3764])

# curve_fit(), calculate the value of a, b in an exponential function
[a, b], res1 = curve_fit(lambda x1,a,b: a*np.exp(b*x1),  x,  y)
# An exponent function is defined as a lambda function lambda x1, a, b: a * numpy.exp (b * x1) 
y1 = a * np.exp(b * x)

# Draw plot using matplotlib.pyplot
plt.plot(x, y, 'b') # draws a graph of original data (blue)
plt.plot(x, y1, 'r') # # draws a graph of approximated data (red)
plt.savefig("data\exponential.png", transparent=True)
plt.show()
```

### Exponential Plot Explained

The example above describes the exponential increase in the spread of the virus. This data can be approximated fairly accurately by an exponential function, at least in pieces along the X-axis.

### When you should use an Exponential plot

Exponential functions can be used to model growth and decay. Exponential functions are ever-increasing so saying that an exponential function models infections growth exactly means it will grow without bounds, no one else to infect. That is not sound reasoning, as the virus is affected by various factors among these are the human population is limited, vaccines, mutations, other viruses. At some point in the future, the number of infections will grow so large that there will not be enough resources (life on the planet) to sustain growth.[20]

## 5.3 Stackplot [18]

Area plots are pretty much similar to the line plot. They are also known as stack plots.

***Sleep tracking app used.***

```python
days = ['Mon 20th', 'Tues 21st', 'Wed 22nd', 'Thurs 23rd', 'Fri 24th']
  
sleeping =[6,7,4,7,5]
eating = [1,2,1,1,3]
college =[14,14,17,15,7]
familytime = [3,1,2,1,9]
  
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='College', linewidth=5)
plt.plot([],[],color='k', label='Family Time', linewidth=5)
 
plt.stackplot(days, sleeping,eating,college,familytime, colors=['m','c','r','k'])
 
plt.xlabel('24 hour period')
plt.ylabel('Day of Christmas week')
plt.title('Stack Plot')
plt.legend()
plt.savefig("data\stackedplot.png", transparent=True)
plt.show()
```

### Stackplot Explained

These plots can be used to track changes over time for two or more related groups that make up one whole category. The example above is the work done during Christmas week into categories, sleeping, eating, college and family time.  

### When you should use a Stackplot

The idea of stack plots is to show “parts to a whole” over time; basically, it’s like a pie-chart, only over time. Stack plots are mainly used to see various trends in variables over a specific period of time. [21]
